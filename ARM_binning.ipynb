```python
import pandas as pd

# Load dataset


# Define custom categories for screen_size column
screen_size_categories = ['small', 'medium', 'large', 'extra-large']
screen_size_bins = [0, 5, 10, 15, 35]
df['screen_size_bin'] = pd.cut(df['screen_size'], bins=screen_size_bins, labels=screen_size_categories)

# Define custom categories for front_column column
front_column_categories = ['short', 'medium', 'long', 'extra-long']
front_column_bins = [0, 10, 20, 30, 35]
df['front_camera_bin'] = pd.cut(df['front_camera_mp'], bins=front_column_bins, labels=front_column_categories)

# Create a new column for release year label
df['release_year_label'] = df['release_year'].apply(lambda x: 'year_' + str(x))

# Select columns to use for association rule mining
transactions = df[['screen_size_bin', 'front_camera_bin', 'release_year_label']]

# Convert to list of lists format
transactions_list = transactions.values.tolist()
```

```python
df.drop(columns=['screen_size', 'front_camera_mp', 'release_year'], inplace=True)
```
